## Завдання 10.

```python
def factorial(n: int) -> int:
    if n < 0:
        return 0
    res = 1
    while True:
        if n in (0, 1):
            break
        res *= n 
        n -= 1
        
    return res
factorial(6)
```

## Завдання 13.

```python
def fibonachi( levels: int, start_num: float = 1) -> float:
    if levels < 2: return 0
    if levels == 2: return start_num
    
    fib_1, fib_2 = start_num, start_num

    for _ in range(2, levels + 1):
        fib_1, fib_2 = fib_2, fib_1 + fib_2


    return max(fib_1, fib_2)
```

```python
print(fibonachi(10))
```

## Завдання 16.

### Напишіть КОРИСТУВАЦЬКУ ФУНКЦІЮ для вибору максимального числа деякої послідовності чисел, які вводить користувач, не використовуючи функцію max

```python
def get_max(*sequence: int | float) -> float | int:
    m = sequence[0]
    for i in sequence:
        if m < i:
            m = i
    return m
```

```python
print(get_max(1, 2, 5, 10))
```

## Завдання 17.

### Напишіть КОРИСТУВАЦЬКУ ФУНКЦІЮ для первірки того, чи належить точка площини з координатами (x,y), деякій прямокутній області.

```python
from typing import Tuple

def is_in_area(dot: Tuple[int, int], area: Tuple[Tuple[int, int], Tuple[int, int]]) -> bool:
    dx, dy = dot
    (x1, y1), (x2, y2) = area
    return (dx >= x1 and dx <= x2 and dy >= y1 and dy <= y2)
```

```python
print(
    is_in_area(
    (100, 2), 
    ((0, 0), (10, 10))
    )
)
```

## Завдання 18.

### Напишіть КОРИСТУВАЦЬКУ ФУНКЦІЮ для перевірки того, існує чи ні трикутних із заданими довжинами сторін (x, y, z). Функція одержує довжини сторін у вигляді аргументів і повертає одно з двох значень 1 або 0. (1 – трикутник існує, 0 – трикутник не існує).

```python
def is_triangle_excist(a: float, b: float, c: float) -> bool:
    if a + b < c or a + c < b or b + c < a:
        return False
    return True
```

```python
print(is_triangle_excist(4, 10, 5))
```

## Завдання 19.

```python
import math
def area(x: float, y: float, z: float) -> float | int:
    if not is_triangle_excist(x, y, z):
        return 0
    p = (x+y+z)/2
    return math.sqrt(p * (p-x) * (p-y) * (p-z))
```

```python
print(area(1,1,1))
```

```python
def func2(C: float) -> int | float:
    if C <= 1:
        return 1
    
    for i in range(2, 101):
        s = 0
        for j in range(1, i):
            s += 1 + (1 / j)
        
        if s >= C:
            return i
        
    return 0
```

```python
print(func2(1000))
```
