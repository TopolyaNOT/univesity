## Задання 10. 
 Порахуйте чистий прибуток (тобто з вирахуванням податків) людини, у випадках, якщо він службовець+ФОП III групи, якщо його дохід (тобто до податків) 4,5к гривень. З’ясуйте в яких випадках краще обрате ту чи іншу групу ФОП. Премії та надбавки: за складність роботи – 1,2к гривень, за стаж – 15%, за інші досягнення – 10%, премія за виконання окремих завдань – 2к гривень, виграв в лотерею – 500 гривень. Виведіть результат у наступному форматі: 

Поточний статус: ...

Дохід: …

Премії: …

Податок на дохід (??%): … 

Військовий збір (??%): …

ЧП: …

(Для зарахування цієї задачі, виконайте математичні розрахунки в Python. Також, поясніть різницю в нарахуванні податків між ФОПами, і чому ІТ-фахівці обирають для себе ІІІ ФОП.) 

```python
# person info
person_status = "soldier"
income = 4500

# premiun and other bonuses
premium = 1200
expirience_bonus = 0.15
other_bonus = 0.1
spesific_bonus = 2000
casino = 500

# tax
pdv = 0.05
military_tax = 0.01
social_tax = 0.22
min_salary = 8647


# result
premium_amount = income * (expirience_bonus + other_bonus) + spesific_bonus
all_income = income + premium + premium_amount + casino 
tax = all_income * (pdv + military_tax) - min_salary * social_tax
net_income = all_income - tax

print(
    f"Person status: {person_status}\n"
    f"Gross income: {all_income:.2f}\n"
    f"Premium: {premium_amount:.2f}\n"
    f"PDV: {pdv:.2f}\n"
    f"Military_tax: {military_tax:.2f}\n"
    f"Total tax: {tax:.2f}\n"
    f"Net income: {net_income:.2f}"
)
```
