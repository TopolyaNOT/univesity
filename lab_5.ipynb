## Завдання 12. 

### Напишіть функцію для обчислення суми елементів одновимірного списку, що ініціалізуються випадковими числами з інтервалу [0;20]. Не викорустовуючи функцію sum або подібні функці

```python
from typing import List

def mega_sum(l: List[int | float]) -> float:
    s = 0
    for i in l: s += i
    return s
```

```python
def rand_list() -> list:
    return [random.random() for _ in range(random.randint(1, 20))]
```

```python
def get_int_list(min_value: int = 1, max_value: int = 20) -> List[int]:
    return [random.randint(min_value, max_value) for _ in range(random.randint(1, 20))]
```

```python
mega_sum(rand_list())
```

```python
def is_ok(l: List[int]) -> bool:
    t = l[0]
    for i in l:
        if t < i:
            return True
        t = i
    return False
```

```python
l7 = []
qa = 0
qb = 0
n = len(l5)
m = len(l6)

for i in range(n + m):
    if qa == n:
        l7.append(l6[qb])
        qb += 1
    elif qb == m:
        l7.append(l5[qa])
        qa += 1
    elif l5[qa] <= l6[qb]:
        l7.append(l5[qa])
        qa += 1
    else:
        l7.append(l6[qb])
        qb += 1

print(l7)
```

## Завдання 19.

```python
def get_matix(min_value: int = -20, max_value: int = 20):
    l = random.randint(1, 20)
    w = random.randint(1, 20)

    return [[random.randint(min_value, max_value) for _ in range(w)] for _ in range(l)]
```

```python
matrix = get_matix()
for ma in matrix: print(ma)
```

```python
def func1(m: List[List[int]]) -> bool:
    for i in range(len(m)):
            try:
                if matrix[i][i] < 0:
                    return True
            except:
                 pass
    return False

print(func1(matrix))
```
