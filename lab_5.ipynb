## Завдання 12. 

### Напишіть функцію для обчислення суми елементів одновимірного списку, що ініціалізуються випадковими числами з інтервалу [0;20]. Не викорустовуючи функцію sum або подібні функці

```python
from typing import List

def mega_sum(l: List[int | float]) -> float:
    s = 0
    for i in l: s += i
    return s

def rand_list() -> list:
    return [random.random() for _ in range(random.randint(1, 20))]
```

```python
def get_int_list() -> List[int]:
    return [random.randint(1, 20) for _ in range(random.randint(1, 20))]
```

```python
mega_sum(rand_list())
```

```python
def above_avg(l: List[int | float]) -> int:
    avg = sum(l) / len(l)
    return [i for i in l if avg < i].__len__()
```

## Завдання 16.

```python
def is_ok(l: List[int]) -> bool:
    t = l[0]
    for i in l:
        if t < i:
            return True
        t = i
    return False
```

```python
print(is_ok(l3))
```
