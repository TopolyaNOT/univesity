## Завдання 10.

```python
def factorial(n: int) -> int:
    if n < 0:
        return 0
    res = 1
    while True:
        if n in (0, 1):
            break
        res *= n 
        n -= 1
        
    return res
factorial(3)
```

## Завдання 12.

```python
def fibonachi( levels: int, start_num: float = 1) -> float:
    if levels < 2:
        return 0
    
    fib_1 = start_num
    fib_2 = start_num

    while levels > 2:
        if fib_1 >= fib_2:
            fib_2 = fib_1 + fib_2
        else:
            fib_1 = fib_1 + fib_2

        levels -= 1


    return max(fib_1, fib_2)
```

```python
print(fibonachi(2, 4))
```
