## Завдання 5. 

### Створіть словник з десяти значень, де ключами будуть числа, а значенням – імена або назви

```python
d = {}
w = [
    "сонце",
    "дерево",
    "небо",
    "кіт",
    "річка",
    "книга",
    "дощ",
    "сміх",
    "вітер",
    "мрія"
]
for i in range(10):
    d[random.randint(0, 100)] = random.choice(w)

print(d)
```
