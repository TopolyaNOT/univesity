## Exircise 3

### Додайте нашій людині приватний атрибут «Кеш». Нехай у нього будуть публічні методи «поповнення», «оплата» та «баланс», які відповідно будуть збільшувати, зменшувати та повертати значення атрибуту

```python
class Human_V3:
    def __init__(self, name: str, sex: str, mood: str):
        self.name = name
        self.sex = sex
        self.mood = mood
        self._cash = 0

    def param(self) -> bool:
        if self.name and self.sex and self.mood:
            end = 'ий' if self.sex.upper() == 'MAN' else 'на'
            print(f'Мене звуть {self.name}. Я - прикольн{end}. У мене {self.mood} настрій')
            return True
        return False
    

    def top_up(self, amount: float) -> bool:
        if  amount <= 0:
            return False
        
        self._cash += amount
        return True
        
    
    def payment(self, amount: float) -> bool:
        if amount <= 0 or self._cash <= 0:
            return False
        
        self._cash -= amount
        return True
    

    def balance(self) -> float:
        return self._cash
```

```python
Monica = Human_V3('Monica', 'wom', 'bad')
Monica.top_up(10_000)
Monica.payment(19999)
print(Monica.balance())
```
